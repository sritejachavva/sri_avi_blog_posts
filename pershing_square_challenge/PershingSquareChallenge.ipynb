```python
import numpy as np
import pandas as pd
import os
from datetime import datetime, timedelta
import yfinance as yf
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import plotly.express as px
import plotly.graph_objects as go
current_directory = os.getcwd()
print(current_directory) 
pd.set_option('display.max_rows', 500)
```

```python
predictions = pd.read_csv('pershing_square_predictions.csv')
predictions['presentation_date'] = pd.to_datetime(predictions['presentation_date'])
predictions.tail(20)
```

```python
predictions['projection_date'] = '';

predictions.loc[predictions['base_3'].notnull(),'projection_date'] = predictions['presentation_date']  + pd.offsets.DateOffset(years=3)
predictions.loc[predictions['base_4'].notnull(),'projection_date'] = predictions['presentation_date']  + pd.offsets.DateOffset(years=4)
predictions.loc[predictions['base_5'].notnull(),'projection_date'] = predictions['presentation_date']  + pd.offsets.DateOffset(years=5)
#predictions.loc[predictions['base_10'].notnull(),'projection_date'] = predictions['presentation_date']  + timedelta(years=10)
predictions['projection_date'] = pd.to_datetime(predictions['projection_date'])
predictions['projection_price'] = predictions.base_3.fillna(predictions.base_4).fillna(predictions.base_5).fillna(predictions.base_10)
predictions.head(10)

```

```python
### create actuals dataframe

def fetch_stock_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data
stock_data = [];

for index, row in predictions.iterrows():
    ticker = row['ticker']
    stock_name = row['stock_name']
    team_rank = row['team_rank']
    start_date = row['presentation_date']
    end_date = row['projection_date']
    reported_presentation_day_price = row['presentation_price']
    presentation_year = row['competition_year']


    data_temp = fetch_stock_data(ticker, start_date, end_date)
    data_temp['ticker'] = ticker
    data_temp['stock_name'] = stock_name
    data_temp['team_rank'] = team_rank
    data_temp['presentation_date'] = start_date
    data_temp['projection_date'] = end_date
    data_temp['presentation_year'] = presentation_year
    data_temp['reported_presentation_day_price'] = reported_presentation_day_price

    
    
    stock_data.append(data_temp)

    
stock_data = pd.concat(stock_data)
stock_data = stock_data.reset_index()
#for stock in predictions['ticker']:
#    print(stock)
```

```python
stock_data['estimate_df'] = 0
stock_data['presentation_day'] = 0
stock_data['projection_day'] = 0


stock_data.loc[((stock_data['presentation_date'] == stock_data['Date']) | (stock_data['projection_date'] == stock_data['Date'])) ,'estimate_df'] = 1
stock_data.loc[((stock_data['presentation_date'] == stock_data['Date'])) ,'presentation_day'] = 1
stock_data.loc[((stock_data['projection_date'] == stock_data['Date'])) ,'projection_day'] = 1


stock_data
```

```python
# Step 1: Find the price of each stock on the first day
first_day_prices = stock_data.groupby(['stock_name','presentation_date'])['Close'].first().reset_index()
first_day_prices.columns = ['stock_name','presentation_date', 'first_day_price']


# Step 2: Merge the first day prices back to the original DataFrame and calculate the ratio
df = stock_data.merge(first_day_prices, on=['stock_name','presentation_date'], how='left')
df['price_ratio'] = df['Close'] / df['first_day_price']

# Display the updated DataFrame
df.head(50)

```

```python
df.to_csv("daily_stock_price.csv")
```
